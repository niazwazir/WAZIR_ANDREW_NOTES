```
from keras.layers import Input, Conv2D, Dense
from keras.models import Model
```

```
img_shape = (39,39,3)
input_img = Input(shape=(img_shape))
C1 = Conv2D(10,(3,3),strides=1,padding='VALID',name='CONV1')(input_img)
C2 = Conv2D(20,(5,5),strides=2,padding='VALID',name='CONV2')(C1)
C3 = Conv2D(40,(5,5),strides=2,padding='VALID',name='CONV3')(C2)
model = Model(input_img, C3)
model.summary()
```
