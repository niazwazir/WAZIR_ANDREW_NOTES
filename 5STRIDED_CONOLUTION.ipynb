**STRIDED CONVOLUTION**
It is the  distance between spatial locations where the convolution kernel is applied. 
In the default scenario, the distance is 1 in each dimension. 
This is also the default value in Tensor Flow.

```
from keras.models import Sequential
from keras.layers import Dense, Conv2D
```

```
model_valid = Sequential([
    Dense(16, input_shape=(7,7,1)),             #IMAGE INPUT SIZE IS 7 x 7   
    Conv2D(32, kernel_size=(3,3),strides=2),    #KERNEL SIZE IS 3 x 3, Stride = 2
       
])
```
