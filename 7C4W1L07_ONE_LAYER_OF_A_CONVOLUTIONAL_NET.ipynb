```

img_shape = (6,6,3)
input_img = Input(shape=(img_shape))
C1 = Conv2D(2,(3,3),padding='VALID',name='CONV1')(input_img)
model = Model(input_img, C1)
opt = optimizers.Adam(lr=0.0003)
model.compile(optimizer=opt,loss='mean_squared_error')
model.summary()
```
