```
model_valid = Sequential([
    Dense(16, input_shape=(6,6,3)),                  #IMAGE SIZE IS 6 x 6, n = 6
    Conv2D(32, kernel_size=(3,3),padding='same'), #KERNEL SIZE IS 3 x 3,, f = 3, Padding = 1, P = 1 OUTPUT = (n +2p - f + 1)x(n + 2p - f + 1) = (6 +2*1-3+ 1) x (6 +2*1-3+ 1) = 6 x 6
       
])
```

```
model_valid.summary() # OUT IS PRESERVED OUTSIZE IS SAME AS 6 x 6
```

```
model_valid = Sequential([
    Dense(16, input_shape=(6,6,3)),                  #IMAGE SIZE IS 6 x 6, n = 6
    Conv2D(32, kernel_size=(3,3),padding='valid'), #KERNEL SIZE IS 3 x 3,, f = 3, No Padding = 0, P = 0 OUTPUT = (n +2p - f + 1)x(n + 2p - f + 1) = (6 + 0-3+ 1) x (6 + 0 -3+ 1) = 4 x 4
    ])
```

```
model_valid.summary() # OUT IS PRESERVED OUTSIZE IS SAME AS 6 x 6
```
