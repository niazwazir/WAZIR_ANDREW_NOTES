```
from IPython.display import Image
Image('Figures/Vertical_Edge.png')
```

```
ucanalytics_logo= cv2.imread('TestImage/baby_GT.bmp')
ucanalytics_logo = cv2.cvtColor(ucanalytics_logo, cv2.COLOR_BGR2RGB)
```

```
%pylab inline
pylab.rcParams['figure.figsize'] = (10, 6)
plt.imshow(ucanalytics_logo)
```

```
#CREATE VERTICAL EDGE DETECTOR
VerticalEdge = np.array([[1, 0, -1],
                 [1, 0, -1],
                 [1, 0, -1]])
```

```
%pylab inline
pylab.rcParams['figure.figsize'] = (12, 8)

#plt.subplot(1, 4, 1)
plt.imshow(VerticalEdge)
plt.title('VerticalEdge')
```

```
%pylab inline
pylab.rcParams['figure.figsize'] = (10, 6)
ucanalytics_logo_flt1 = cv2.filter2D(ucanalytics_logo, -1, VerticalEdge)
plt.imshow(ucanalytics_logo_flt1)
```
